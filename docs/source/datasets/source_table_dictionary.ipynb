# Data Source Table Dictionary/Key

The source table provides the information needed to create sources and load data as well as background information about each dataset. Below are the definitions for each column in the source table:

| Column | Definition | Used to Create: | Required? |
| ------- | ------- | ------- | ------- |
| **State** | Name of State | `Source` | NO, only needed if datasets exists for same **SourceName** in different states |
| **SourceName** | Name of source (typically city/town or shortened name of a police department (PD)) | `Source` | YES |
| **Agency**  | Typically same as **SourceName** but can be 'MULTIPLE'if dataset contains data for multiple agencies  | `load`* | NO, if 'MULTIPLE', agency input to `load` can be used to filter for a specifc agency (PD) |
| **AgencyFull** | Full name of agency | | |
| **TableType** | Type of data ('TRAFFIC STOPS', 'USE OF FORCE', etc.) | `load`* | YES |
| **coverage_start** | Start date of dataset | `load`* | YES if requesting date range of dataset, first value of date range in `date` input should be *greater than* or equal to this date |
| **coverage_end** | Stop date of dataset (at last check) | `load`* | YES if requesting date range of dataset, last value of date range in `date` input should *generally* be *less than* or equal to this date. Data beyond this date may be available if the data has been updated since last check. |
| **last_coverage_check** | Date that `coverage_start` and `coverage_end` were last updated |  | |
| **Year** | Year of the dataset or 'MULTIPLE' for multi-year or 'NONE' if the data is not for particular year(s) | `load`* | YES if requesting entire dataset, set to `date` to this value |
| **agency_originated** | Whether data was originally collected by the agency (Empty indicates that data was collected AND released by agency) | | |
| **supplying_entity** | organization that supplied the data if not the agency described by the data | | |
| **Description** | Description of datasets |  | |
| **source_url** | Homepage URL for dataset |  |  |
| **readme** | URL for data dictionary (if exists) | 
| **URL** | URL of data file or API endpoint. If **dataset_id** is not empty, URL is combined with **dataset_id** to locate data. |
| **DataType** | Type of data (CSV, Excel, ArcGIS, Socrata, etc.) |
| **date_field** | Date column of dataset |
| **dataset_id** | If exists, dataset ID(s) used in combination with **URL** to locate data |
| **agency_field** | For multi-agency data, agency column of dataset |
| **min_version** | Minimum OPD version required to load a dataset |
| **py_min_version** | Minimum Python version required to load a dataset |
| **query** | Query to perform on the data such that returned data is of type **TableType**. Access the full dataset by going to the **source_url** |
