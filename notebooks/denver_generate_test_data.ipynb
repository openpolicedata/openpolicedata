```python
import requests
import pandas as pd
import geopandas as gpd
import numbers
import numpy as np
import math

import re
def extract_year(str_date):
    pattern = '\d\d-\d\d-(\d\d\d\d)'
    if match := re.search(pattern, str_date, re.IGNORECASE):
        year = match.group(1)
    else:
        year=None
    return year
```
